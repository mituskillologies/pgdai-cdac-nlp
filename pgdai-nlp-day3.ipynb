```python
from nltk.corpus import wordnet
```

```python
synset = wordnet.synsets("company")
print('Word and Type : ' + synset[0].name())
print('Synonym : ' + synset[0].lemmas()[0].name())
print('The meaning of the word : ' + synset[0].definition())
print('Example : ' + str(synset[0].examples()))
```

```python
syn = []
ant = []
for synset in wordnet.synsets("later"):
    for lemma in synset.lemmas():
        syn.append(lemma.name())    #add the synonyms
        if lemma.antonyms():    #When antonyms are available, add them into the list
            ant.append(lemma.antonyms()[0].name())
print('Synonyms: ' ,set(syn))
print('Antonyms: ' ,set(ant))
```

```python
word = 'set'
print("VERB : ", wordnet.synsets(word, pos = wordnet.VERB))
print("ADJECTIVE : ", wordnet.synsets(word, pos = wordnet.ADJ))
print("NOUN : ", wordnet.synsets(word, pos = wordnet.NOUN))
```

```python
word = 'jam'
syn = wordnet.synsets(word, pos = wordnet.VERB)
for x in syn:
    print(x,":", x.definition())
```

```python
from nltk.corpus import stopwords
swords = stopwords.words('english')

from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()

from nltk import FreqDist
from nltk import pos_tag
```

```python
html = urllib.request.urlopen('https://en.wikipedia.org/wiki/Jupiter')

data = html.read()

soup = BeautifulSoup(data,'lxml')
data = soup.get_text()

tokens = word_tokenize(data)

tokens1 = [x.lower() for x in tokens if x.isalpha() or x.isdigit()]
tags = pos_tag(tokens1)
tokens2 = [x[0] for x in tags if x[1].startswith('N')]

tokens3 = [wnl.lemmatize(x,'n') for x in tokens2 if x not in swords]

freq = FreqDist(tokens3)
freq.plot(10, cumulative=False)
```

```python
from collections import Counter
x = dict(Counter(tokens3))
```

```python
{k: v for k, v in sorted(x.items(), key=lambda item: item[1], reverse=True)}
```
