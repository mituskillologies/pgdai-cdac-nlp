```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
list(enumerate(alphabets))
```

```python
char_to_int = dict((c, i) for i, c in enumerate(alphabets))
```

```python
int_to_char = dict((i, c) for i, c in enumerate(alphabets))
```

```python
seq_length = 1
```

```python
x = []
y = []
for i in range(len(alphabets) - seq_length):
  seq_in = alphabets[i:i + seq_length]
  seq_out = alphabets[i + seq_length]
  x.append([char_to_int[char] for char in seq_in])
  y.append(char_to_int[seq_out])
  print(seq_in,'->',seq_out)
```

```python
y
```

```python
import numpy as np
```

```python
x = np.reshape(x, (len(x), seq_length, 1))
```

```python
x.shape
```

```python
# Feature scaling
x = x / 25
```

```python
from keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model = Sequential()

model.add(LSTM(32, input_shape=(x.shape[1], x.shape[2])))

model.add(Dense(26, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', 
              optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
model.fit(x, y, epochs=500, batch_size=1)
```

```python
model.evaluate(x, y, batch_size=1)
```

```python
new = 'R'
new = char_to_int[new]
dx = np.reshape(new, (1, len([new]),1))
dx = dx / 25
pred = model.predict(dx, verbose=False)
print(int_to_char[np.argmax(pred)])
```
